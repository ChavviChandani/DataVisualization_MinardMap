```
survivor_df
```

#### Plotting Chart of Army that Advanced and Returned, i.e Survivours

```
army= alt.Chart(survivor_df).mark_trail().encode(
    longitude='LONP',
    latitude='LATP',
    size=alt.Size(
        'SURV',
        legend=None,
        scale=alt.Scale(range=[1,55])
         ),
         detail= 'DIV',
    color=alt.Color(
        'DIR',
        scale=alt.Scale(
            domain=['A','R'],
            range=['#EBD2A8', '#888888']
        ),
        legend=None
      )
    )
```

```
army
```

#### Poltting the "number" of people

```
number_of_people=mindard_df.iloc[:,8:]
```

```
number_of_people.drop_duplicates('SURV',keep='first',inplace=True)
```

```
number_of_people["LONP"] += 0.18 * (number_of_people["DIV"])
number_of_people["LATP"] += number_of_people["DIR"].replace({"A": 0.30, "R": -0.18})
```

```
number_of_people_chart = alt.Chart(number_of_people).mark_text(
    fontStyle='italic',fontSize=8,angle=280).encode(longitude='LONP:Q',latitude='LATP:Q',text='SURV')
```

```
number_of_people_chart
```

#### Plotting the name of the cities

```
name_of_cities_chart = alt.Chart(cities_df).mark_text(
    fontStyle='italic',fontSize=8,angle=360).encode(longitude='LONC:Q',latitude='LATC:Q',text='CITY')
```

```
name_of_cities_chart
```

#### Plotting Temperature Chart

```
temperature_df=temperature_df.fillna("")
```

```
temperature_df['TEXT']=temperature_df['TEMP'].astype(str)+'°'+temperature_df['MON'].astype(str)
```

```
temperature_chart_1 = alt.Chart(temperature_df).mark_line(
    color="#000000"
).encode(
    x=alt.X('LONT:Q',axis=None),
    y=alt.Y('TEMP', axis=alt.Axis(title='Temperature',orient='right'))
)
```

```
temperature_chart_1
```

```
temperature_chart_2 = alt.Chart(temperature_df).mark_text( 
    dx=5,
    dy=20, 
).encode(
    x=alt.X('LONT:Q',axis=None),
    y=alt.Y('TEMP', axis=alt.Axis(title='Temperature',orient='right',grid=True)),
    text="TEXT"
)
```

```
temperature_chart_2
```

```
temperature_chart=temperature_chart_1 + temperature_chart_2
```

```
temperature_chart
```

```
temperature_chart = temperature_chart.properties(height=100)
```

```
temperature_chart
```

#### Plotting Army, number_of_people and name_of_cities together

```
minards_map=army + number_of_people_chart + name_of_cities_chart
```

```
minards_map
```

```
minards_map = alt.vconcat(minards_map, temperature_chart).configure_view(
    width=900,height=300,strokeWidth=0,).configure_axis(
    grid=False)
```
